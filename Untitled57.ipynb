```python
# Cell 1: Import libraries
# Cell 1：匯入必要函式庫
import numpy as np
import cv2
import os
from tqdm.notebook import tqdm
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import spsolve
import matplotlib.pyplot as plt
from skimage import color
from skimage.color import deltaE_ciede2000
from sewar.full_ref import vifp
```

```python
# Cell 2: Dark channel computation
#Cell 2：定義暗通道函式
def dark_channel(img, patch_size=15):
    """Compute the dark channel of an RGB image."""
    h, w = img.shape[:2]
    pad = patch_size // 2
    min_rgb = np.min(img, axis=2)
    dark = np.zeros((h, w), dtype=min_rgb.dtype)
    for i in range(h):
        for j in range(w):
            i_min, i_max = max(0, i-pad), min(h, i+pad+1)
            j_min, j_max = max(0, j-pad), min(w, j+pad+1)
            dark[i, j] = np.min(min_rgb[i_min:i_max, j_min:j_max])
    return dark
```

```python
# Cell 3: Atmospheric light estimation
#Cell 3：大氣光估計
def estimate_atmospheric_light(img, dark, top_percent=0.001):
    """Estimate atmospheric light from brightest pixels in the dark channel."""
    num_pixels = max(int(dark.size * top_percent), 1)
    flat_dark = dark.flatten()
    indices = np.argsort(flat_dark)[-num_pixels:]
    h, w = dark.shape
    y = indices // w
    x = indices % w
    brightest = img[y, x]
    A = np.max(brightest, axis=0)
    return A
```

```python
# Cell 4: Initial transmission estimation
#Cell 4：初始透射率估計
def estimate_transmission(img, A, omega=0.95, patch_size=15):
    """Estimate the coarse transmission map."""
    normalized = img / A.reshape(1,1,3)
    dark_norm = dark_channel(normalized, patch_size)
    transmission = 1 - omega * dark_norm
    return transmission
```

```python
# Cell 5: Guided filter for transmission refinement
#Cell 5：Guided Filter 與 Soft Matting
def guided_filter(I, p, r, eps):
    """Guided filter implementation."""
    mean_I = cv2.boxFilter(I, cv2.CV_64F, (r, r))
    mean_p = cv2.boxFilter(p, cv2.CV_64F, (r, r))
    cov_Ip = cv2.boxFilter(I*p, cv2.CV_64F, (r, r)) - mean_I*mean_p
    var_I = cv2.boxFilter(I*I, cv2.CV_64F, (r, r)) - mean_I*mean_I
    a = cov_Ip / (var_I + eps)
    b = mean_p - a*mean_I
    mean_a = cv2.boxFilter(a, cv2.CV_64F, (r, r))
    mean_b = cv2.boxFilter(b, cv2.CV_64F, (r, r))
    return mean_a * I + mean_b

# Soft matting refinement
def soft_matting(img, transmission, lambda_=1e-4, eps=1e-10):
    """Refine transmission with soft matting (sparse linear system)."""
    h, w = transmission.shape
    N = h * w
    gray = cv2.cvtColor((img*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)/255.0
    dx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    dy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    mean_dx = cv2.boxFilter(dx, cv2.CV_64F, (3,3))
    mean_dy = cv2.boxFilter(dy, cv2.CV_64F, (3,3))
    var_dx = cv2.boxFilter(dx*dx, cv2.CV_64F, (3,3)) - mean_dx*mean_dx
    var_dy = cv2.boxFilter(dy*dy, cv2.CV_64F, (3,3)) - mean_dy*mean_dy

    rows, cols, data = [], [], []
    for y in range(h):
        for x in range(w):
            idx = y*w + x
            rows.append(idx); cols.append(idx); data.append(lambda_)
            for ny in range(max(0,y-1), min(h,y+2)):
                for nx in range(max(0,x-1), min(w,x+2)):
                    if ny==y and nx==x: continue
                    nidx = ny*w + nx
                    weight = 1.0/(eps + var_dx[y,x] + var_dy[y,x])
                    rows.append(idx); cols.append(nidx); data.append(-weight)
                    data[rows.index(idx)] += weight

    L = csr_matrix((data, (rows, cols)), shape=(N, N))
    refined = spsolve(L, lambda_*transmission.flatten())
    return refined.reshape(h, w).clip(0,1)
```

```python
# Cell 6: Wrap refinement methods
#Cell 6：整合透射率優化
def refine_transmission(img, transmission, method='guided', r=40, eps=1e-6, lambda_=1e-4):
    """Choose refinement method."""
    I_gray = cv2.cvtColor((img*255).astype(np.uint8), cv2.COLOR_RGB2GRAY)/255.0
    if method == 'guided':
        return guided_filter(I_gray, transmission, r, eps)
    elif method == 'soft_matting':
        return soft_matting(img, transmission, lambda_, eps)
    else:
        raise ValueError("Unknown method")
```

```python
# Cell 7: Recover dehazed image
#Cell 7：影像復原
def recover_image(img, A, transmission, t0=0.1):
    """Recover scene radiance."""
    t = np.maximum(transmission, t0)[..., np.newaxis]
    J = (img - A) / t + A
    return np.clip(J, 0, 1)
```

```python
# Cell 9: Batch processing function
#Cell 9：批量處理函式與執行
def batch_dehaze(input_dir, output_dir, **params):
    os.makedirs(output_dir, exist_ok=True)
    files = [f for f in os.listdir(input_dir) if f.lower().endswith(('.jpg','.png','.jpeg'))]
    for fname in tqdm(files, desc='Batch Dehazing'):
        in_path = os.path.join(input_dir, fname)
        out_path = os.path.join(output_dir, fname)
        # single-image pipeline
        img_bgr = cv2.imread(in_path)
        img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)/255.0
        dark = dark_channel(img, params.get('patch_size',15))
        A = estimate_atmospheric_light(img, dark)
        t0 = estimate_transmission(img, A, params.get('omega',0.95), params.get('patch_size',15))
        t1 = refine_transmission(img, t0, method=params.get('method','guided'))
        J = recover_image(img, A, t1, params.get('t0',0.1))
        # save outputs
        cv2.imwrite(out_path, cv2.cvtColor((J*255).astype(np.uint8), cv2.COLOR_RGB2BGR))
```

```python
# Cell 1: Imports
import os
import numpy as np
import cv2
import torch
import lpips
from tqdm.notebook import tqdm
from image_similarity_measures.quality_metrics import fsim as ism_fsim
from skimage.metrics import structural_similarity as structural_ssim
from skimage import color
import sewar

# Cell 2: Image loader
def load_img(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img.astype(np.float32) / 255.0

# Cell 3: PSNR & SSIM
def compute_psnr(img1, img2):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return float('inf')
    return 20 * np.log10(1.0 / np.sqrt(mse))

def compute_ssim(img1, img2):
    """
    Compute SSIM between two RGB images in [0,1].
    """
    return structural_ssim(img1, img2, data_range=1.0, channel_axis=2)

# Cell 4: CIEDE2000 & VIF
from skimage.color import deltaE_ciede2000
from sewar.full_ref import vifp

def compute_ciede2000(img1, img2):
    lab1 = color.rgb2lab(img1)
    lab2 = color.rgb2lab(img2)
    delta_e = deltaE_ciede2000(lab1, lab2)
    return np.mean(delta_e)

def compute_vif(img1, img2):
    img1_gray = cv2.cvtColor((img1*255).astype(np.uint8),
                             cv2.COLOR_RGB2GRAY).astype(np.float64)/255.0
    img2_gray = cv2.cvtColor((img2*255).astype(np.uint8),
                             cv2.COLOR_RGB2GRAY).astype(np.float64)/255.0
    return vifp(img1_gray, img2_gray)

# …（后续 evaluate_pair 和 batch_evaluate 保持不变）…



# ——— Cell 5: 更新 evaluate_pair 加入新指標 ———

def evaluate_pair(dehazed_path, gt_path):
    """
    Given file paths, compute PSNR, SSIM, LPIPS, CIEDE2000, and VIF.
    """
    dehazed = load_img(dehazed_path)
    gt      = load_img(gt_path)

    # 先計算 PSNR、SSIM
    psnr_val  = compute_psnr(gt, dehazed)
    ssim_val  = compute_ssim(gt, dehazed)

    # 新增 CIEDE2000 色差
    ciede_val = compute_ciede2000(gt, dehazed)
    # 新增 VIF
    vif_val   = compute_vif(gt, dehazed)

    # 再計算 LPIPS
    dehazed_t = torch.from_numpy(dehazed).permute(2,0,1).unsqueeze(0)
    gt_t      = torch.from_numpy(gt).permute(2,0,1).unsqueeze(0)
    lpips_fn  = lpips.LPIPS(net='alex')
    lpips_val = lpips_fn(dehazed_t, gt_t).item()

    return psnr_val, ssim_val, lpips_val, ciede_val, vif_val

# ——— Cell 6: 更新 batch_evaluate 累計並列印新指標 ———

def batch_evaluate(dehazed_dir, gt_dir):
    """
    Iterate through dehazed images folder, match GT, and print metrics including
    PSNR, SSIM, LPIPS, CIEDE2000, and VIF.
    """
    sums  = {'psnr':0.0, 'ssim':0.0, 'lpips':0.0,
             'ciede':0.0, 'vif':0.0}
    count = 0

    files = [f for f in os.listdir(dehazed_dir)
             if f.lower().endswith(('.jpg','.png','.jpeg'))]
    for fname in tqdm(files, desc='Evaluating'):
        if fname.endswith(('_transmission.png','_visualization.png')):
            continue

        dehazed_p = os.path.join(dehazed_dir, fname)
        gt_fname  = fname.replace('_hazy','_GT')
        gt_p      = os.path.join(gt_dir, gt_fname)
        if not os.path.exists(gt_p):
            print(f"GT not found for {fname}, skipping.")
            continue

        # 解包所有五個指標
        psnr_v, ssim_v, lpips_v, ciede_v, vif_v = evaluate_pair(dehazed_p, gt_p)

        sums['psnr']  += psnr_v
        sums['ssim']  += ssim_v
        sums['lpips'] += lpips_v
        sums['ciede'] += ciede_v
        sums['vif']   += vif_v
        count += 1

        print(f"{fname} → "
              f"PSNR: {psnr_v:.2f}, "
              f"SSIM: {ssim_v:.4f}, "
              f"LPIPS: {lpips_v:.4f}, "
              f"CIEDE2000: {ciede_v:.2f}, "
              f"VIF: {vif_v:.4f}")

    if count > 0:
        print("\nAverage PSNR:       {:.2f}".format(sums['psnr']/count))
        print("Average SSIM:       {:.4f}".format(sums['ssim']/count))
        print("Average LPIPS:      {:.4f}".format(sums['lpips']/count))
        print("Average CIEDE2000:  {:.2f}".format(sums['ciede']/count))
        print("Average VIF:        {:.4f}".format(sums['vif']/count))
    else:
        print("No valid images to evaluate.")


# Cell 7: Run batch evaluation
dehazed_folder = r"C:\Users\tonyg\Dense_Haze_NTIRE19\dark"
gt_folder      = r"C:\Users\tonyg\Dense_Haze_NTIRE19\GT"
batch_evaluate(dehazed_folder, gt_folder)
```
